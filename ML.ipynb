```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
def nonlin(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))
```

```python
np.random.seed(4298576)
```

```python
XY = pd.read_csv("data/winequality-white1.csv")
XY = XY.sample(n = 10)
XY = (XY-XY.min())/(XY.max()-XY.min())
X = XY.drop('quality',axis=1).values
Y = np.asmatrix(XY['quality'].values)
```

```python
syn0 = 2*np.random.random((11,10)) - 1
syn1 = 2*np.random.random((10,1)) - 1
```

```python
for iter in range(100):
    l0 = X
    l1 = nonlin(np.dot(l0,syn0))
    l2 = nonlin(np.dot(l1,syn1))
    
    l2_error = Y - l2
    if ((iter%100) == 0):        
        print('Error '+ str(np.mean(np.abs(l2_error))))
    
    l2_delta = l2_error * nonlin(l2,deriv=True)
    
    l1_error = np.dot(l2_delta, syn1.T)
    
    l1_delta = l1_error * nonlin(l1,True)
    
    syn1 += np.dot(l1.T,l2_delta)
    syn0 += np.dot(l0.T,l1_delta)
```
